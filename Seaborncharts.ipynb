```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("flights")
var
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=var, x="year", y="passengers", hue="month")
plt.title("Passenger Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show
```

```python
pivot_var = var.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_var, annot=True, cmap="YlGnBu")
plt.title("Passenger Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=var, x="month", y="passengers", palette="viridis")
plt.title("Average Passengers by Month")
plt.xlabel("Month")
plt.ylabel("Average Passengers")
plt.show()
```
